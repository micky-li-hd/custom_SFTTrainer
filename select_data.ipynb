```python
import os

# 
tar_dir = "/mnt/v-haodongli/cot_output_test"
log_dir = "/mnt/v-haodongli/status"

#  tar 
failed_log_files = []

#  .log 
log_files = [f for f in os.listdir(log_dir) if f.endswith('.log')]

for log_file in log_files:
    base_name = os.path.splitext(log_file)[0]  #  .log "00001"
    log_file_path = os.path.join(log_dir, log_file)

    #  .tar 
    tar_file = f"{base_name}.tar"
    tar_file_path = os.path.join(tar_dir, tar_file)

    #  log  "failed"
    try:
        with open(log_file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            if "failed" in content.lower():
                failed_log_files.append(tar_file_path)
    except Exception as e:
        print(f"Error reading {log_file_path}: {e}")

# 
print("\nFailed log files:")
for f in failed_log_files:
    print(f)
```
